```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
```

```python
driver =webdriver.Chrome('chromedriver.exe')
```

```python
HotelName=[]
Rating=[]
Address=[]
Price=[]
Review=[]
Facilities=[]
CoupleFriendly=[]
MoreOptions=[]
```

```python
driver.get('https://www.makemytrip.com/hotels/hotel-listing/?checkin=12022020&city=CTDEL&checkout=12032020&roomStayQualifier=2e0e&locusId=CTDEL&country=IN&locusType=city&searchText=Delhi&visitorId=86bcec80-4336-44b8-bdaa-45b6ad73ac74')
```

```python
content = driver.page_source
soup = BeautifulSoup(content)
#print(soup)
```

```python
names=soup.findAll('p',attrs={'class':'latoBlack font22 blackText appendBottom12'})
print(len(names))
```

```python
for n in names:
    HotelName.append(n.text)
```

```python
rating=soup.findAll('p',attrs={'class':'font11 capText darkText'})
print(len(rating))
```

```python
for n in rating:
    Rating.append(n.text)
```

```python
location=soup.findAll('p',attrs={'itemprop':'address'})
print(len(location))
```

```python
for n in location:
    Address.append(n.text)
```

```python
price=soup.findAll('p',attrs={'class':'latoBlack font26 blackText appendBottom5'})
print(len(price))
```

```python
for n in price :
    Price.append(n.text)
```

```python
facilities=soup.findAll('ul',attrs={'class':'amenList darkText'})
print(len(facilities))
```

```python
for n in facilities :
    Facilities.append(n.text)
    
```

```python
review=soup.findAll('span',attrs={'class':'ratingText latoRegular font11'})
print(len(review))
```

```python
for n in review :
    Review.append(n.text)
```

```python
cuplefriendly=soup.findAll('ul',attrs={'class':'tags font11 latoBold appendBottom5 darkText'})
print(len(cuplefriendly))
```

```python
for n in cuplefriendly :
    CoupleFriendly.append(n.text)
```

```python
breakfast=soup.findAll('p',attrs={'class':'font12 appendBottom10'})
print(len(breakfast))
```

```python
for n in breakfast:
    MoreOptions.append(n.text)
```

```python
Datalist={'HotelName':HotelName,'Address':Address,'Prices':Price,'Rating':Rating,'Review':Review,'CoupleFriendly':CoupleFriendly,'Facilities':Facilities,'MoreOptions':MoreOptions}
```

```python
dfDelhi=pd.DataFrame.from_dict(Datalist, orient='index')
dfDelhi=dfDelhi.transpose()
```
